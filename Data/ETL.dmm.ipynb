```python
import pandas as pd
import datetime as dt
```

```python
#Import Found csv
found_df = pd.read_csv("AAC_Found.csv")

#Select rows where Type= "Dog"
found_df = found_df.loc[found_df['Type'] == 'Dog']

#Select columns ('Animal ID', 'Found Location', 'At AAC', 'Intake Date', 'Looks Like', 'Color','Sex','Age','Image Link')
found_df = found_df[['Animal ID', 
                     'Found Location', 
                     'At AAC', 
                     'Intake Date',
                     'Looks Like', 
                     'Color',
                     'Sex',
                     'Age',
                     'Image Link']]

#Rename looks like column
found_df.rename(columns = {'Looks Like': 'Breed(unconfirmed)'}, inplace = True)

#Split Sex column into gender and found sex
#split value of Sex upon intake column
found_sex_split = found_df['Sex'].str.split(" ", n = 1, expand = True) 
  
# making separate intake sex column from new data frame 
found_df["Found_Sex"]= found_sex_split[0] 

# making separate gender column from new data frame 
found_df["Gender"]= found_sex_split[1] 
  
# Dropping old Sex upon intake column
found_df.drop(columns =['Sex'], inplace = True) 

#Split Found Location column into address and coordinates
#split value of found location column
found_loc_split = found_df['Found Location'].str.split("(", n = 1, expand = True) 
  
# making separate address column from new data frame 
address = found_loc_split[0].str.rstrip()

#Extract zip code (last 5 characters of address)
found_df["Zip_code"] = address.str[-5:]

#split Lat/Long on ","
found_coord_split = found_loc_split[1].str.split(",", n = 1, expand = True) 
# making separate coordinate columns from new data frame 
found_df["Latitude"] = found_coord_split[0] 
found_df["Longitude"] = found_coord_split[1].str.replace(')', '') 


found_df.head()

#found_df.count()
```

```python
#Import Intakes csv
intakes_df = pd.read_csv("AAC_Intake.csv")

#Select rows where Type= "Dog"
intakes_df = intakes_df.loc[intakes_df['Animal Type'] == 'Dog']

#Select columns ('Animal ID','Name','DateTime','Found Location','Intake Type','Intake Condition','Sex upon Intake','Age upon Intake','Breed','Color')
intakes_df = intakes_df[['Animal ID',
                        'DateTime',
                        'Found Location',
                        'Intake Type',
                        'Intake Condition',
                        'Sex upon Intake',
                        'Age upon Intake',
                        'Breed',
                        'Color']]

#Rename datetime column
intakes_df.rename(columns = {'DateTime': 'Intake_Date'}, inplace = True)

##Split Sex Upon intake column into gender and intake sex
#split value of Sex upon intake column
intake_sex_split = intakes_df['Sex upon Intake'].str.split(" ", n = 1, expand = True) 
  
# making separate intske sex column from new data frame 
intakes_df["Intake_Sex"]= intake_sex_split[0] 
  
# making separate gender column from new data frame 
intakes_df["Gender"]= intake_sex_split[1] 
  
# Dropping old Sex upon intake column
intakes_df.drop(columns =['Sex upon Intake'], inplace = True) 


##Split Found location column into address, city, state
#split value of location column on (
intake_loc_split = intakes_df['Found Location'].str.split("(", n = 1, expand = True) 
  
# making separate state column from new data frame
intakes_df["Found State"] = intake_loc_split[1].str.replace(')', '')

#split intake_loc_split[0] on " in "
intake_street_split = intake_loc_split[0].str.split(" in ", n = 1, expand = True) 
  
#Sort by Animal ID, then by intake date
intakes_df.sort_values(by=["Animal ID", "Intake_Date"], ascending=[True, True], inplace=True)

intakes_df.head()
```

```python
##Split Found location column into street, city, state
found_street = []
found_city = []

#Code below is an attempt to sort street vs. city values correctly
#Loop through intake_loc_split[0] values
for string in intake_loc_split[0]:
    #If " in " exists in string then split street/city on " in "
    if " in " in string:
        #split strong on " in "
        intake_street_split = string.split(" in ", 1) 

        # append 
        found_street.append(intake_street_split[0])
        found_city.append(intake_street_split[1])

    #If not then place string in "found city"
    else:
        found_street.append('none')
        found_city.append(string)
        
#add city and street columns to intakes-df
intakes_df["Found City"] = found_city
intakes_df["Found Street"] = found_street

#create date index list
date_ID = []

#check if the current Animal ID is the same as the 1 before it
match = intakes_df['Animal ID'].eq(intakes_df['Animal ID'].shift())

#loop through df
for i in match:
    #if the current Animal ID is the same as the 1 before it...
    if i == True:
        #add 1 to dateid
        dateid += 1
    #if not...
    else:
        #start dateid as 1
        dateid = 1
    #append dateid value to date_ID list
    date_ID.append(dateid)
    
#add date_ID to df
intakes_df['Date_ID'] = date_ID

intakes_df.head(20)
```

```python
#Add found location to found_df
found_loc_df = found_df.merge(intakes_df, how='inner', on='Animal ID')

#reorder columns
found_loc_df = found_loc_df[['Animal ID', 
                             'Found Street',
                           'Found City',
                           'Found State',
                             'Zip_code',
                             'Latitude',
                             'Longitude',
                             'At AAC', 
                             'Intake Date',
                             'Breed(unconfirmed)', 
                             'Color_x',
                             'Found_Sex',
                             'Gender_x',
                             'Age',
                             'Image Link']]

found_loc_df.head()
#found_loc_df.count()

#found_loc_df['At AAC'].value_counts()

#to csv
found_loc_df.to_csv('AAC_Found_Final.csv', index= False)
```

```python
#Import Outcomes csv
outcomes_df = pd.read_csv("AAC_Outcome.csv")

#Select rows where Type= "Dog"
outcomes_df = outcomes_df.loc[outcomes_df['Animal Type'] == 'Dog']


#Select columns ('Animal ID', 'Name', 'DateTime', 'Date of Birth', 'Outcome Type', 'Outcome Subtype', 'Sex upon Outcome', 'Age upon Outcome', 'Breed', 'Color')
outcomes_df = outcomes_df[['Animal ID', 
                          'Name', 
                          'DateTime', 
                          'Date of Birth', 
                          'Outcome Type', 
                          'Sex upon Outcome', 
                          'Age upon Outcome']]

#Rename datetime column
outcomes_df.rename(columns = {'DateTime': 'Outcome_Date'}, inplace = True)

#Split Sex column into gender and outcome sex. Remove Gender column
#split value of Sex upon outcome column
outcome_sex_split = outcomes_df['Sex upon Outcome'].str.split(" ", n = 1, expand = True) 
  
# making separate sex outcome column from new data frame 
outcomes_df["Outcome_Sex"]= outcome_sex_split[0] 
  
# Dropping old sex upon outcome column
outcomes_df.drop(columns =['Sex upon Outcome'], inplace = True) 

#Remove * from Names
outcomes_df['Name'] = outcomes_df['Name'].str.replace('*','')

#Sort by Animal ID, then by outcome date
outcomes_df.sort_values(by=["Animal ID", "Outcome_Date"], ascending=[True, True], inplace=True)

#create date index list
date_ID = []

#check if the current Animal ID is the same as the 1 before it
match = outcomes_df['Animal ID'].eq(outcomes_df['Animal ID'].shift())

#loop through df
for i in match:
    #if the current Animal ID is the same as the 1 before it...
    if i == True:
        #add 1 to dateid
        dateid += 1
    #if not...
    else:
        #start dateid as 1
        dateid = 1
    #append dateid value to date_ID list
    date_ID.append(dateid)
    
#add date_ID to df
outcomes_df['Date_ID'] = date_ID

outcomes_df.head()
#outcomes_df['Animal ID'].value_counts()
#outcomes_df.info()
```

```python
#Merge DataFrames (intake, outcome, found?)
merged_df = pd.merge(intakes_df, outcomes_df, how='inner', on=['Animal ID','Date_ID'])

##merge (or zip) each row w/ condition
#merge this instance of Animal ID(intake) with the next same instance of animal id(outcome)
#if theres not an equal ID, move to the next 1

#Split color by "/" (color 1, color2, color3)
#split value of color column
merged_color_split = merged_df['Color'].str.split("/", n = 2, expand = True) 
  
# making separate color1 column from new data frame 
merged_df["Color1"]= merged_color_split[0] 

# making separate color2 column from new data frame 
merged_df["Color2"]= merged_color_split[1] 

# making separate color3 column from new data frame 
#merged_df["Color3"]= merged_color_split[2]
  
# Dropping old color column
merged_df.drop(columns =['Color'], inplace = True) 

#create breed lists
breed1 =[]
breed2 = []

#Split breed (breed1, breed2, breed3)
for breed in merged_df['Breed']:
    if "/" in breed:
        #split value of breed column
        merged_breed_split = breed.split("/", 2) 

        # making separate breed1 column from new data frame 
        breed1.append(merged_breed_split[0])

        # making separate breed2 column from new data frame 
        breed2.append(merged_breed_split[1])
        
    elif "Mix" in breed: 
        # making separate breed1 column from new data frame 
        breed1.append(breed[:-3])

        # making separate breed2 column from new data frame 
        breed2.append(breed[-3:])
    
    else:
        # making separate breed1 column from new data frame 
        breed1.append(breed)

        # making separate breed2 column from new data frame 
        breed2.append("none")

# making separate breed3 column from new data frame 
#merged_df["Breed3"] = merged_breed_split[2]

#add breed1 and breed2 columns to df
merged_df['Breed1'] = breed1
merged_df['Breed2'] = breed2
  
# Dropping old breed column
merged_df.drop(columns = ['Breed'], inplace = True) 


#Create Duration of Stay column
#convert columns to datetime type
merged_df["Outcome_Date"] = pd.to_datetime(merged_df["Outcome_Date"])
merged_df["Intake_Date"] = pd.to_datetime(merged_df["Intake_Date"])

#subtract intake and outcome dates to form duration column
merged_df["AAC_Duration"] = (merged_df["Outcome_Date"] - merged_df["Intake_Date"])


#Remove NaN rows or replace NaN values
merged_df['Name'].fillna("none", inplace = True)
merged_df['Color2'].fillna("none", inplace = True)
merged_df.dropna(inplace=True)

#Remove rows with negative duration
merged_df1 = merged_df[merged_df.AAC_Duration > dt.timedelta(minutes=0)]
merged_df1 = merged_df[merged_df.AAC_Duration > dt.timedelta(days=0)]

#Reorder columns
merged_df1 = merged_df[['Animal ID', 
                      'Name', 
                      'Date of Birth',
                       'Gender',
                       'Color1',
                       'Color2',
                       'Breed1',
                       'Breed2',
                       'Found Street',
                       'Found City',
                       'Found State',
                       'Intake_Date',
                       'Intake Type',
                       'Intake Condition',
                       'Age upon Intake',
                       'Intake_Sex',
                       'Outcome_Date',
                      'Outcome Type', 
                      'Age upon Outcome',
                       'Outcome_Sex', 
                       'AAC_Duration']]

merged_df1.head(15)
```
