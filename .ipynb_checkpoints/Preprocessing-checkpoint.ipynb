```python
import pandas as pd
import numpy as np
from datetime import datetime as dt
from datetime import timedelta as td
```

```python
df = pd.read_csv("./Data/AAC_In_Out_Final.csv")
df.head()
#list(df.columns)
#print(df.dtypes)
#df.Breed2.unique()
```

```python
# drop unecessary columns
dog_df1 = df[["Gender", "Color1", "Color2", "Breed1", "Breed2", "Intake_Date", "Intake Type", "Intake Condition", "Intake_Sex", "Outcome_Date", "Outcome Type", "Outcome_Age_Months", "Outcome_Sex", "AAC_Duration"]]
dog_df1.head()
```

```python
# Get min and max of Outcome_Age_Months column
dog_df1["Outcome_Age_Months"].max()
dog_df1["Outcome_Age_Months"].min()

# Drop AAC_Duration, and Outcome_Age_Months rows with negative values
dog_df2 = dog_df1[dog_df1["Outcome_Age_Months"] >=0]
```

```python
#convert Outcome date column to month
type(dog_df2.Outcome_Date[0])
dog_df2.Outcome_Date = pd.to_datetime(dog_df2.Outcome_Date)
dog_df2['Outcome_Date_Month'] = dog_df2.Outcome_Date.dt.month

#print result
dog_df2.head()
```

```python
def td_cleaner2(timed):
    seconds = timed.total_seconds()
    return(seconds)
```

```python
#convert Intake date column to month
type(dog_df2.Intake_Date[0])
dog_df2.Intake_Date = pd.to_datetime(dog_df2.Intake_Date)
dog_df2['Intake_Date_Month'] = dog_df2.Intake_Date.dt.month

dog_df2.head()
```

```python


dog_df2["AAC_Duration_Clean"] = pd.to_timedelta(dog_df2["AAC_Duration"])
```

```python
dog_df2["TDSeconds"] = dog_df2.AAC_Duration_Clean.apply(td_cleaner2)

```

```python
dog_df2["TDSeconds"]

dog_df2 = dog_df2[dog_df2["TDSeconds"] >=0]

dog_df2
```

```python
dog_df2["TDSeconds"] = pd.to_numeric(dog_df2.TDSeconds)


dog_df2["TDDays"] = dog_df2["TDSeconds"]/86400

dog_df2.head()
```

```python
#convert Intake date column to month
type(dog_df2.Intake_Date[0])
dog_df2.Intake_Date = pd.to_datetime(dog_df2.Intake_Date)
dog_df2['Intake_Date_Month'] = dog_df2.Intake_Date.dt.month

dog_df2.head()
```

```python
#convert Outcome date column to month
type(dog_df2.Outcome_Date[0])
dog_df2.Outcome_Date = pd.to_datetime(dog_df2.Outcome_Date)
dog_df2['Outcome_Date_Month'] = dog_df2.Outcome_Date.dt.month

#print result
dog_df2.head()
```

```python
dog_df2 = dog_df2.loc[:,~dog_df2.columns.duplicated()]


dog_cleandf = dog_df2[['Gender', 'Color1', 'Color2', 'Breed1', 'Breed2', 'Intake Type', 'Intake Condition', 'Intake_Sex', 'Outcome Type', 'Outcome_Age_Months', 'Outcome_Sex', 'Outcome_Date_Month', 'Intake_Date_Month', 'TDDays']]
dog_cleandf.head()
```

```python
dog_cleandf.columns = ['Gender', 'Color1', 'Color2', 'Breed1', 'Breed2', 'Intake_Type', 'Intake_Condition', 'Intake_Sex', 'Outcome_Type', 'Outcome_Age_Month', 'Outcome_Sex', 'Outcome_Date_Month', 'Intake_Date_Month', 'Duration_Day']

dog_cleandf.info()
```

```python
dog_cleandf.to_csv('MachineLearning.csv', index=False)
```
